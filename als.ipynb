```
from pyspark.sql import SparkSession
```

```
session = SparkSession.builder.appName("ALS").getOrCreate()
```

```
ratingDf = session.read.csv('/FileStore/tables/ratings-8.csv',header = "true",inferSchema = "true")
```

```
bookDf = session.read.csv('/FileStore/tables/books-6.csv',header="true",inferSchema = "true")
```

```
from pyspark.ml.recommendation import ALS
from pyspark.ml.evaluation import RegressionEvaluator
```

```
als = ALS(userCol = "user_id", itemCol = "book_id", ratingCol = "rating", coldStartStrategy="drop", nonnegative=True, implicitPrefs=False)
```

```
(training , test) = ratingDf.randomSplit([0.8 , 0.2])
```

```
model = als.fit(training)
```

```
predictions = model.transform(test)
```

```
predictions.show()
```

```
evaluatr = RegressionEvaluator(metricName="rmse",labelCol="rating",predictionCol="prediction")
```

```
rmse = evaluatr.evaluate(predictions)
```

```
print(rmse)
```
